```python
dfMethan = wb.data.DataFrame('EN.ATM.METH.KT.CE')
dfMethan
```

# Replace country codes by country names

```python
# List of country codes
country_codes = dfMethan.index
country_codes
```

```python
# Retrieve country names if it exists.
country_names = []
for code in country_codes:
    country = pycountry.countries.get(alpha_3=code)
    if country:
        country_names.append(country.name)
    else:
        country_names.append('Country name not found')

print(country_names)
```

```python
# Function to get country name from code
def get_country_name(code):
    country = pycountry.countries.get(alpha_3=code)
    if country:
        return country.name
    else:
        if code == 'WLD':
            return 'World'
        return code
        
# Replace index with country names
dfMethan.index = dfMethan.index.map(get_country_name)

dfMethan.head()
```

```python
# Function to filter out non-3-letter indices
def filter_valid_codes(code):
    return len(code) == 3

# Filter DataFrame based on valid indices
valid_indices = dfMethan.index[dfMethan.index.map(filter_valid_codes)]

# Create new DataFrame with filtered rows
df_filtered = dfMethan.loc[valid_indices]

# Display filtered DataFrame
df_filtered
```

```python
# Remove filtered rows from original DataFrame
dfMethan.drop(index=df_filtered.index, inplace=True)
dfMethan.head()
```

# Missing values

```python
missing_values = dfMethan.isnull().sum()
missing_values
```

```python
# I want to display columns with 215 missing values

# Filter to display only columns with missing values
columns_empty = missing_values[missing_values == 216]

# Display the columns with missing values and their counts
columns_empty.sum()/216
```

```python
# Let's remove our empty columns
dfMethan.transpose()
dfMethan = dfMethan.drop(columns=columns_empty.index)
```

```python
# Liste des pays à récupérer
european_top_10 = ['Netherlands', 'Poland', 'Ireland', 'Italy', 'France', 
                   'Spain', 'Germany', 'Belgium', 'Austria', 'Sweden']

# Initialiser une liste pour stocker les indices des pays européens
indices_pays_europeens = []

# Parcourir la liste des pays européens
for pays in european_top_10:
    # Obtenir l'index du pays dans le DataFrame dfMethan
    index_pays = dfMethan.index.get_loc(pays)
    indices_pays_europeens.append(index_pays)

# Récupérer les lignes correspondant aux pays européens
pays_europeens = dfMethan.iloc[indices_pays_europeens]

# Afficher les pays récupérés
print("Pays récupérés:")
pays_europeens
```

# Remove 'YR' from columns names

```python
# Remove 'YR' prefix from column names
pays_europeens = pays_europeens.rename(columns={col: col[2:] for col in pays_europeens.columns if col.startswith('YR')})
pays_europeens
```

```python
# Remove 'YR' prefix from column names
dfMethan = dfMethan.rename(columns={col: col[2:] for col in dfMethan.columns if col.startswith('YR')})
dfMethan
```

```python
# Plotting
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed

for country in pays_europeens.index:
    plt.plot(pays_europeens.columns, pays_europeens.loc[country], marker='o', label=country)

plt.title('Methan Emissions Over Time for Top 10 European Countries')
plt.xlabel('Year')
plt.ylabel('Methan Emissions (kT)')
plt.legend()
plt.grid(True)
plt.xticks(pays_europeens.columns)  # Set the x-axis ticks to be the years
plt.xticks(rotation=45)
plt.tight_layout()

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# Assuming pays_europeens is your DataFrame with countries in the index

for country in pays_europeens.index:
    plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
    
    # Extract the data for the current country
    country_data = pays_europeens.loc[country]
    years = country_data.index.values.astype(int)
    emissions = country_data.values
    
    # Plot the methane emissions over time
    plt.plot(years, emissions, marker='o', label='Methane Emissions')
    
    # Compute the mean methane emissions per year
    mean_emissions = np.mean(emissions)
    
    # Fit a linear regression model
    X = years.reshape(-1, 1)  # Reshape for sklearn
    y = emissions
    model = LinearRegression().fit(X, y)
    trend_line = model.predict(X)
    
    # Plot the linear regression line
    plt.plot(years, trend_line, label='Linear Regression', color='red')
    
    plt.title(f'Methane Emissions Over Time for {country}')
    plt.xlabel('Year')
    plt.ylabel('Methane Emissions')
    plt.grid(True)
    plt.xticks(years[::5])  # Set the x-axis ticks to be every 5 years
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.legend()
    
    # Show plot
    plt.show()
```

```python
# Create an empty DataFrame to store the transformed rows
rows_list = []

# Iterate through each cell in the DataFrame
for country, row in dfMethan.iterrows():
    for year, Methan in row.items():
        # Create a dictionary for each row
        row_dict = {
            'Country': country,
            'Date': year,
            'MethanEmission (kt)': Methan
        }
        # Append the dictionary to the list
        rows_list.append(row_dict)

# Create a new DataFrame from the list of dictionaries
df_transformed = pd.DataFrame(rows_list)

# Display the resulting transformed DataFrame
print("Transformed DataFrame:")
df_transformed.head()
```
