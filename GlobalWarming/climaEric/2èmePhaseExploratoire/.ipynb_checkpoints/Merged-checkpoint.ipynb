```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, HTML
```

```python
import pandas as pd

# Load the datasets
df_co2 = pd.read_csv('CO2.csv')
df_population = pd.read_csv('PopulationTotalClean.csv')
df_methan = pd.read_csv('Methan.csv')
df_gdp_growth = pd.read_csv('GDPGrowth.csv')
df_astc = pd.read_csv('ASTC.csv')
```

```python
print("CO2 DataFrame Summary:")
df_co2.head()
df_co2.info()

print("\nPopulation DataFrame Summary:")
df_population.head()
df_population.info()

print("\nMethan DataFrame Summary:")
df_methan.head()
df_methan.info()

print("\nGDP Growth DataFrame Summary:")
df_gdp_growth.head()
df_gdp_growth.info()

print("\nASTC DataFrame Summary:")
df_astc.head()
df_astc.info()
```

```python
print("\nUnique countries in CO2 DataFrame:", df_co2['Country'].unique())
print("\nUnique countries in Population DataFrame:", df_population['Country'].unique())
print("\nUnique countries in Methan DataFrame:", df_methan['Country'].unique())
print("\nUnique countries in GDP Growth DataFrame:", df_gdp_growth['Country'].unique())
print("\nUnique countries in ASTC DataFrame:", df_astc['Country'].unique())

print("\nUnique dates in CO2 DataFrame:", df_co2['Date'].unique())
print("\nUnique dates in Population DataFrame:", df_population['Date'].unique())
print("\nUnique dates in Methan DataFrame:", df_methan['Date'].unique())
print("\nUnique dates in GDP Growth DataFrame:", df_gdp_growth['Date'].unique())
print("\nUnique dates in ASTC DataFrame:", df_astc['Date'].unique())
```

```python
# Reset index to ensure all DataFrames have the same structure
df_co2.reset_index(inplace=True)
df_population.reset_index(inplace=True)
df_methan.reset_index(inplace=True)
df_gdp_growth.reset_index(inplace=True)
df_astc.reset_index(inplace=True)

# Concatenate the DataFrames along 'Country' and 'Date'
df_merged = pd.concat([df_co2, df_population, df_methan, df_gdp_growth, df_astc], axis=1, join='inner')

# Display the merged DataFrame
df_merged
```

```python
# Sélection des colonnes requises
selected_columns = ['CO2Emission (kt)', 'MethanEmission (kt)', 'Population', 'GDPGrowth', 'ASTC']
df_selected = df_merged[selected_columns]

# Calcul de la matrice de corrélation
correlation_matrix = df_selected.corr()

# Affichage de la matrice de corrélation
print("Matrice de Corrélation :")
correlation_matrix
```

```python
# Calculate correlation matrix
correlation_matrix = df_selected[['Population', 'CO2Emission (kt)', 'MethanEmission (kt)']].corr()

# Create a heatmap using Seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```

# Inner Merged Dataframes

```python
# Merge the two DataFrames on 'Country' and 'Date'
inner_merged_df = pd.merge(df_co2, df_methan, on=['Country', 'Date'], how='inner')

# Merge the two DataFrames on 'Country' and 'Date'
inner_merged_df = pd.merge(inner_merged_df, df_population, on=['Country', 'Date'], how='inner')

# Remove ['Unnamed: 0'] column 
inner_merged_df.drop(columns=['Unnamed: 0'], inplace=True)

# Merge the two DataFrames on 'Country' and 'Date'
inner_merged_df = pd.merge(inner_merged_df, df_gdp_growth , on=['Country', 'Date'], how='inner')

# Display the merged DataFrame
print("Merged DataFrame:")
inner_merged_df.columns
```

```python
inner_merged_df.drop(columns=['Unnamed: 0_x', 'Unnamed: 0_y', 'Unnamed: 0'], inplace=True)
```

```python
# Calculate correlation matrix
correlation_matrix = inner_merged_df[['Population', 'CO2Emission (kt)', 'MethanEmission (kt)', 'GDPGrowth']].corr()

# Create a heatmap using Seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```
