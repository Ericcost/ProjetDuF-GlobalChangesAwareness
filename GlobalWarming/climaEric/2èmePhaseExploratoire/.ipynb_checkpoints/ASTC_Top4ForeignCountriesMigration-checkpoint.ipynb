# Replace country codes by country names

```python
# Set the 'ISO3' column as the index
dftc_t.set_index('ISO3', inplace=True)
```

```python
# List of country codes
country_codes = dftc_t.index
country_codes
```

```python
# Retrieve country names if it exists.
country_names = []
for code in country_codes:
    country = pycountry.countries.get(alpha_3=code)
    if country:
        country_names.append(country.name)
    else:
        country_names.append('Country name not found')

print(country_names)
```

```python
# Function to get country name from code
def get_country_name(code):
    country = pycountry.countries.get(alpha_3=code)
    if country:
        return country.name
    else:
        if code == 'WLD':
            return 'World'
        return code
        
# Replace index with country names
dftc_t.index = dftc_t.index.map(get_country_name)

dftc_t.head()
```

```python
# Function to filter out non-3-letter indices
def filter_valid_codes(code):
    return len(code) == 3

# Filter DataFrame based on valid indices
valid_indices = dftc_t.index[dftc_t.index.map(filter_valid_codes)]

# Create new DataFrame with filtered rows
df_filtered = dftc_t.loc[valid_indices]

# Display filtered DataFrame
df_filtered
```

```python
# Remove filtered rows from original DataFrame
dftc_t.drop(index=df_filtered.index, inplace=True)
dftc_t.head()
```

```python
# Function to find the oldest non-null value in a row
def replace_nan_with_oldest(row):
    oldest_value = None
    for value in row:
        if pd.notna(value):
            if oldest_value is None or value < oldest_value:
                oldest_value = value
    return row.fillna(oldest_value)

# Apply the function row-wise to replace NaN values
df_t = df_t.apply(replace_nan_with_oldest, axis=1)

# Display the cleaned DataFrame
df_t
```

```python
# Liste des pays à récupérer
foreign_top_4 = ['Algeria', 'Morocco', 'Portugal', 'Romania']

# Initialiser une liste pour stocker les indices des pays européens
indices_pays_migratoires = []

# Parcourir la liste des pays européens
for pays in foreign_top_4:
    # Obtenir l'index du pays dans le DataFrame df_t
    index_pays = df_t.index.get_loc(pays)
    indices_pays_migratoires.append(index_pays)

# Récupérer les lignes correspondant aux pays européens
pays_migratoires = df_t.iloc[indices_pays_migratoires]

# Afficher les pays récupérés
print("Pays récupérés:")
pays_migratoires
```

# Remove 'F' from columns names

```python
# Remove 'F' prefix from column names
pays_migratoires = pays_migratoires.rename(columns={col: col[1:] for col in pays_migratoires.columns if col.startswith('F')})
pays_migratoires
```

```python
# Remove 'F' prefix from column names
df_t = df_t.rename(columns={col: col[1:] for col in df_t.columns if col.startswith('F')})
df_t
```

```python
# Plotting
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed

for country in pays_migratoires.index:
    plt.plot(pays_migratoires.columns, pays_migratoires.loc[country], marker='o', label=country)

plt.title('ASTC Over Time for Top 10 European Countries')
plt.xlabel('Year')
plt.ylabel('ASTC')
plt.legend()
plt.grid(True)
plt.xticks(pays_migratoires.columns)  # Set the x-axis ticks to be the years
plt.xticks(rotation=45)
plt.tight_layout()

# Show plot
plt.show()
```

```python
import numpy as np
from scipy.stats import linregress

for country in pays_migratoires.index:
    plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
    plt.plot(pays_migratoires.columns, pays_migratoires.loc[country], marker='o', label='ASTC')
    
    # Calculer les paramètres de la régression linéaire
    x = np.arange(len(pays_migratoires.columns))  # Années en indices
    y = pays_migratoires.loc[country].values
    slope, intercept, r_value, p_value, std_err = linregress(x, y)
    
    # Tracer la ligne de régression
    plt.plot(pays_migratoires.columns, intercept + slope * x, 'r', label=f'Régression linéaire (r={r_value:.2f})')
    
    plt.title(f'ASTC Over Time for {country}')
    plt.xlabel('Year')
    plt.ylabel('ASTC')
    plt.grid(True)
    
    # Configurer les ticks de l'axe des x pour afficher tous les 5 ans
    ticks = np.arange(0, len(pays_migratoires.columns), 5)
    plt.xticks(ticks=ticks, labels=pays_migratoires.columns[ticks], rotation=45)
    
    plt.legend()
    plt.tight_layout()
    
    # Show plot
    plt.show()
```
