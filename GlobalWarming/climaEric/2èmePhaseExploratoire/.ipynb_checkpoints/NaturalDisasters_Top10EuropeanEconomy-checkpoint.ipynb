```python
import pandas as pd
import matplotlib.pyplot as plt
import pycountry
```

# Get Dataset

```python
dfnd = pd.read_csv('./DataSets/NaturalDisasters.csv')
dfnd.columns
```

```python
dfnd
```

```python
values_array = [
    'Number of deaths from drought',
    'Number of people injured from drought',
    'Number of people affected from drought',
    'Number of people left homeless from drought',
    'Number of total people affected by drought',
    'Reconstruction costs from drought',
    'Insured damages against drought',
    'Total economic damages from drought'
]
```

```python
# Dictionary to store DataFrames
df_pivot_dict = {}

# Create and store DataFrames for each value
for value in values_array:
    # Pivot the DataFrame to get the desired format
    df_pivot = dfnd.pivot(index='Country name', columns='Year', values=value)
    
    # Store the DataFrame in the dictionary with a dynamic name
    df_pivot_dict[f"df_pivot_{value}"] = df_pivot
    
    # Display the DataFrame for the current value
    print(f"DataFrame for '{value}':")
    df_pivot
    print("\n")
```

```python
# Example of accessing a specific DataFrame from the dictionary
print("Accessing DataFrame 'Number of deaths from drought':")
df_pivot_dict['df_pivot_Number of deaths from drought']
```

# Let's check missing values

### Template

```python
'Number of deaths from drought',
    'Number of people injured from drought',
    'Number of people affected from drought',
    'Number of people left homeless from drought',
    'Number of total people affected by drought',
    'Reconstruction costs from drought',
    'Insured damages against drought',
    'Total economic damages from drought'
```

```python
df_pivot_dict['df_pivot_Number of deaths from drought'].info()
df_pivot_dict['df_pivot_Number of deaths from drought'].shape
missing_values = df_pivot_dict['df_pivot_Number of deaths from drought'].isnull().sum()
df_pivot_dict['df_pivot_Number of deaths from drought'].isnull().sum().sum()
# Liste des pays à récupérer
european_top_10 = ['Netherlands', 'Poland', 'Ireland', 'Italy', 'France', 
                   'Spain', 'Germany', 'Belgium', 'Austria', 'Sweden']

# Initialiser une liste pour stocker les indices des pays européens
indices_pays_europeens = []

# Parcourir la liste des pays européens
for pays in european_top_10:
    # Obtenir l'index du pays dans le DataFrame df_pivot_dict['df_pivot_Number of deaths from drought']
    index_pays = df_pivot_dict['df_pivot_Number of deaths from drought'].index.get_loc(pays)
    indices_pays_europeens.append(index_pays)

# Récupérer les lignes correspondant aux pays européens
pays_europeens = df_pivot_dict['df_pivot_Number of deaths from drought'].iloc[indices_pays_europeens]

# Afficher les pays récupérés
print("Pays récupérés:")
pays_europeens
```

# Plot

```python
# Plotting
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed

for country in pays_europeens.index:
    plt.plot(pays_europeens.columns, pays_europeens.loc[country], marker='o', label=country)

plt.title('ASTC Over Time for Top 10 European Countries')
plt.xlabel('Year')
plt.ylabel('ASTC')
plt.legend()
plt.grid(True)
plt.xticks(pays_europeens.columns)  # Set the x-axis ticks to be the years
plt.xticks(rotation=45)
plt.tight_layout()

# Show plot
plt.show()
```
